# Atari DQN Training Notebook

This notebook combines training scripts for DQN on CartPole, MsPacman, and SpaceInvaders Atari games.

```python
import os
os.environ["GYM_DISABLE_ENV_CHECKER"] = "1"

import gymnasium as gym
import ale_py
from stable_baselines3 import DQN
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.atari_wrappers import AtariWrapper

MODEL_DIR = "models"
os.makedirs(MODEL_DIR, exist_ok=True)
```

## CartPole Training

```python
def train_cartpole(total_timesteps=100_000):
    env = gym.make("CartPole-v1")
    model = DQN("MlpPolicy", env, verbose=1, tensorboard_log="./tb_cartpole")
    model.learn(total_timesteps=total_timesteps)
    model_path = os.path.join(MODEL_DIR, "cartpole_dqn")
    model.save(model_path)
    mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=20)
    print(f"CartPole evaluation: mean_reward={mean_reward:.2f} +/- {std_reward:.2f}")
    env.close()
```

```python
train_cartpole()
```

## MsPacman Training

```python
def train_mspacman(total_timesteps=400_000):
    env = gym.make("ALE/MsPacman-v5")
    env = AtariWrapper(env)
    model = DQN("CnnPolicy", env, verbose=1, tensorboard_log="./tb_mspacman")
    model.learn(total_timesteps=total_timesteps)
    model_path = os.path.join(MODEL_DIR, "mspacman_dqn")
    model.save(model_path)
    mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
    print(f"MsPacman evaluation: mean_reward={mean_reward:.2f} +/- {std_reward:.2f}")
    env.close()
```

```python
train_mspacman()
```

## SpaceInvaders Training

```python
def train_spaceinvaders(total_timesteps=400_000):
    env = gym.make("SpaceInvadersNoFrameskip-v4")
    env = AtariWrapper(env)
    model = DQN("CnnPolicy", env, verbose=1, tensorboard_log="./tb_spaceinvaders")
    model.learn(total_timesteps=total_timesteps)
    model_path = os.path.join(MODEL_DIR, "spaceinvaders_dqn")
    model.save(model_path)
    mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
    print(f"SpaceInvaders evaluation: mean_reward={mean_reward:.2f} +/- {std_reward:.2f}")
    env.close()
```

```python
train_spaceinvaders(10_000)
```
